```python
import os
path = 'C:/Users/cicer/Documents/ThesisWork/Ocean/FileIDing/'
coupls = 'C:/Users/cicer/Documents/ThesisWork/Ocean/Couplings/'
```

```python
def sum_absolute_couplings(filename):
    total_sum = 0.0
    
    with open(filename, 'r') as file:
        for line in file:
            # Skip the comments and blank lines
            if line.startswith("#") or line.strip() == "":
                continue
            
            parts = line.split()
            try:
                total_sum += abs(float(parts[2]))  # Sum absolute values of the third column
            except ValueError:
                print(f"Skipping line: {line.strip()}")
    
    return total_sum
```

```python
fileList = os.listdir(coupls)
```

```python
for file in fileList:
    energy = sum_absolute_couplings(coupls+file)
    print(f"{coupls+file}: {energy}")
```
